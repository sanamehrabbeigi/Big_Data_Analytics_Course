```python
%sql
SELECT * FROM `hive_metastore`.`default`.`data_citibike_stations`;
```

```python
df = spark.sql("SELECT * FROM data_citibike_stations")
df_filtered = df.filter("longitude <= -73.94 AND longitude >= -74.04")
count = df_filtered.count()
display(count)
display(df_filtered)
```

```python
df_filtered_region= df.filter("region_id = 71")
total_bikes_available = df_filtered_region.agg({"num_bikes_available": "sum"}).collect()[0][0]
display(total_bikes_available)
display()
```

```python
df_max_cap = df.agg({'capacity': 'max'}).collect()[0][0]
display(df_max_cap)
df_max_cap_id = df.filter(df.capacity == df_max_cap).select("station_id")
display(df_max_cap_id)
```
